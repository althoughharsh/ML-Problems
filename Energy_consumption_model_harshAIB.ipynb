```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train_df = pd.read_csv('/content/train_energy_data.csv')
test_df = pd.read_csv('/content/test_energy_data.csv')
```

```python
print("Training Data:")
print(train_df.head())

print("\nTesting Data:")
print(test_df.head())
```

```python
X = train_df.drop(['Building Type','Day of Week','Energy Consumption'], axis=1)
y = train_df['Energy Consumption']
```

```python
X_test = test_df.drop(['Building Type','Day of Week','Energy Consumption'], axis=1)
y_test = test_df['Energy Consumption']
```

```python
train_df.head()
```

```python
X
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X_test)
```

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(results.head())
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R2 Score:", r2)
```
